```python
import numpy as np # Working with vectors and arrays
import matplotlib 
import matplotlib.pyplot as plt # Standard module for plotting and visualization
import scipy as sp 
import scipy.stats as stats # Scientific algorithms & basic statistics
import pandas as pd # Adds DataFrames to Python (powerful spreadsheets)
import seaborn as sns # Visualization of statistical data
import statsmodels.formula.api as sm # Provides classes and functions for many statistical models
from statsmodels.stats import power # Statistical power test
```

# Task 51a)

```python
myDF = stats.norm(10348,250)
x = np.linspace(10348-1000, 10348+1000, 501)
plt.figure(0)
plt.plot(x,myDF.pdf(x)) # plot domain x against y-axis being the pdf
plt.title('Pdf of normally distributed X with mean 10348, std 250')
```

# Task 51b) and c)

```python
(mu,n,sigma) = (10348, 100, 2500)
# Part a
mean = mu
std_error = sigma/np.sqrt(n)

# Part b
vary = 200
(value_1,value_2) = (mu-vary, mu+vary)
(z_1,z_2) = ((value_1-mu)/std_error,(value_2-mu)/std_error)
print('The probability the sample mean will be within plus/minus 200 of the population',
      f'mean is {stats.norm.cdf(z_2) - stats.norm.cdf(z_1)}')
# Part c
threshold = 12000
z_threshold = (threshold - mu)/std_error
print(f'The probability the sample mean will be greater than 12000 is {1-stats.norm.cdf(z_threshold)}') 
```

# Task 58a)

```python
(pop_por,n) = (0.15, 240)
mean = pop_por 
std_error = np.sqrt(pop_por*(1-pop_por)/n)

myDF = stats.norm(mean,std_error)
x = np.linspace(mean-0.10, mean+0.10)
plt.figure(0)
plt.plot(x,myDF.pdf(x)) # plot domain x against y-axis being the pdf
plt.title('Pdf of normally distributed Pbar with mean 0.15, std 0.023')
```
