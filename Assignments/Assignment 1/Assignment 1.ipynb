# Task 24e

```python
data = 'CompactSUV.xlsx'
data = pd.read_excel(data)
display(data)

a = 0
b = 0
c = 0
d = 0
e = 0

#Frequency distribution
for i in data['Owner Satisfaction']:
    if i == '--':
        a += 1
    elif i == '-':
        b += 1
    elif i == 0:
        c += 1
    elif i == '+':
        d += 1
    elif i == '++':
        e += 1

#Percentage Frequency        
per_a = a/15*100
per_b = b/15*100
per_c = c/15*100
per_d = d/15*100
per_e = e/15*100


#Plotting bar chart
x = ['--', '-', '0', '+', '++']
y = [per_a, per_b, per_c, per_d, per_e]

plt.bar(x,y)
plt.title('Bar Chart for Owner Satisfaction')
plt.xlabel('Owner Satisfaction')
plt.ylabel('Percentage Frequency')
plt.show()

```

```python
data = 'CompactSUV.xlsx'
data = pd.read_excel(data)
display(data)

a = 0
b = 0
c = 0
d = 0

#Frequency distribution
for i in data['Acceleration (0-60) Sec']:
    if i >= 7.0 and i < 8.0:
        a += 1
    elif i >= 8.0 and i < 9.0:
        b += 1
    elif i >= 9.0 and i < 10.0:
        c += 1
    elif i >= 10.0 and i < 11.0:
        d += 1
print(a,b,c,d)        


#Plotting histogram
x = data['Acceleration (0-60) Sec']
plt.title('Histogram for Acceleration')
plt.xlabel('Acceleration')
plt.ylabel('Frequency')
plt.hist(x, bins = 4)
plt.show()

        
```

# Task 48a and b

```python
data = 'BBB.xlsx'
data = pd.read_excel(data)
display(data)

cable = 0
car = 0
cell = 0
collection = 0
bank = 0


#Frequency
for i in data['Industry']:
    if i == 'Cable':
        cable += 1
        cable_per = cable/200*100
    elif i == 'Car':
        car += 1
        car_per = car/200*100
    elif i == 'Cell':
        cell += 1
        cell_per = cell/200*100
    elif i == 'Collection':
        collection += 1
        collection_per = collection/200*100
    elif i == 'Bank':
        bank += 1
        bank_per = bank/200*100
        
data2 = pd.DataFrame({'Industry': ['Cable', 'Car', 'Cell', 'Collection', 'Bank'],
        'Frequency': [cable, car, cell, collection, bank],
        'Percent Frequency':[cable_per, car_per, cell_per, collection_per, bank_per]
        })
display(data2)


#Plotting bar chart
x = ['Cable', 'Car', 'Cell', 'Collection', 'Bank']
y = [cable_per, car_per, cell_per, collection_per, bank_per]

plt.bar(x,y)
plt.title('Bar Chart for Consumer Complaints')
plt.xlabel('Industry')
plt.ylabel('Percentage Frequency')
plt.show()
```

# Task 57a

```python
data = 'ElectricVehicles.xlsx'
data = pd.read_excel(data)
display(data)

first = data[2013]
second = data[2015]

n=5
r = np.arange(n)
width = 0.25
  
  
plt.bar(r, first, color = 'b',
        width = width, edgecolor = 'black',
        label='2013')
plt.bar(r + width, second, color = 'g',
        width = width, edgecolor = 'black',
        label='2015')
  
plt.xlabel("Region")
plt.ylabel("Number of electric plug-in vehicle sales")
plt.title("Number of electric plug-in vehicle sales in each year")

plt.xticks(r + width/2,['China','Western Europe','U.S.','Japan', 'Canada'])
plt.legend()
  
plt.show()
```

# Task 57b

```python
data = 'ElectricVehicles.xlsx'
data = pd.read_excel(data)
display(data)

first = data[2013]
second = data[2015]
region = data['Region']

for i in range(0,5):
    if region[i] == 'China':
        ch13 = first[i] / (first[i] + second[i]) * 100
        ch15 = second[i] / (first[i] + second[i]) * 100
    elif region[i] == 'Western Europe':
        eu13 = first[i] / (first[i] + second[i]) * 100
        eu15 = second[i] / (first[i] + second[i]) * 100
    elif region[i] == 'United States':
        us13 = first[i] / (first[i] + second[i]) * 100
        us15 = second[i] / (first[i] + second[i]) * 100
    elif region[i] == 'Japan':
        ja13 = first[i] / (first[i] + second[i]) * 100
        ja15 = second[i] / (first[i] + second[i]) * 100
    elif region[i] == 'Canada':
        ca13 = first[i] / (first[i] + second[i]) * 100
        ca15 = second[i] / (first[i] + second[i]) * 100
    

x = ['China','Western Europe','U.S.','Japan', 'Canada']
y1 = [ch13, eu13, us13, ja13, ca13]
y2 = [ch15, eu15, us15, ja15, ca15]

print(y1)

plt.bar(x, y1, color='r', width = 0.4)
plt.bar(x, y2, bottom=y1, color='b', width = 0.4)

plt.xlabel("Region")
plt.ylabel("Percentage allocation")
plt.title("Percentage allocation of electric plug-in vehicle sales for 2013 and 2015")


plt.show()
```
